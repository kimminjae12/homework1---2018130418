```python
import numpy as np

# 6자리의 특정 수
target = '123456'

# 시도 횟수
num_trials = 500

# 출연횟수를 저장할 변수
count = 0

for i in range(num_trials):
    # 백만개의 정수 난수 생성
    random_integers = np.random.randint(low=0, high=10, size=1000000)
    
    # 6개씩 묶어서 검사
    for j in range(len(random_integers) - 5):
        six_numbers = ''.join([str(num) for num in random_integers[j:j+6]])
        if six_numbers == target:
            count += 1
            break
print("이어진 수가 나타난 횟수: ", count)
probability = count / num_trials
print("이어진 수가 나타날 확률: ", probability)
```

```python
import numpy as np

# 행렬 생성
c = np.array([[1,9,9,9,0,8,1,3], [2,0,1,8,1,3,0,4]])
# c+7
c_plus_7 = c + 7
# c-7
c_minus_7 = c - 7
# c*7
c_times_7 = c * 7
# c+c
c_plus_c = c + c
# c/c
c_divide_c = c / c
c_transpose = np.transpose(c)
c_times_c = "오류"
c_times_c_transpose = np.dot(c, c_transpose)
c_transpose_times_c = np.dot(c_transpose, c)
print("c = \n",c)
print("c+7 = \n",c_plus_7)
print("c-7 = \n",c_minus_7)
print("c*7 = \n",c_times_7)
print("c+c = \n",c_plus_c)
print("c/c = \n",c_divide_c)
print("c·c = \n",c_times_c)
print("c^T·c = \n",c_transpose_times_c)
print("c·c^T = \n",c_times_c_transpose)
```

```python
#첫번째 예시
class Student:
    def __init__(self, name, department, year):
        self.name = name
        self.department = department
        self.year = year
        self.grade = {}

    def enter_grade(self, subject, grade):
        self.grade[subject] = grade

    def get_grade(self, subject):
        if subject in self.grade:
            return self.grade[subject]
        else:
            return "No grade available for the subject."

    def calc_avg_grade(self):
        if len(self.grade) == 0:
            return "No grade available."
        else:
            return sum(self.grade.values()) / len(self.grade)

# 학생 객체 생성
s = Student("홍길동", "컴퓨터공학과", 2)

# 성적 입력
s.enter_grade("수학", 90)
s.enter_grade("영어", 80)
s.enter_grade("국어", 85)

# 성적 조회
print(s.get_grade("수학"))
print(s.get_grade("과학"))

# 성적 평균 계산
print(s.calc_avg_grade())

#두번째 예시
class Fish:
    def __init__(self, name, species, length, color, habitat):
        self.name = name
        self.species = species
        self.length = length
        self.color = color
        self.habitat = habitat

    def swim(self):
        print(f"{self.name} is swimming.")

    def eat(self, food):
        print(f"{self.name} is eating {food}.")

    def change_color(self, new_color):
        self.color = new_color
        print(f"{self.name} changed color to {new_color}.")


nemo = Fish("Nemo", "Clownfish", 3, "Orange", "Coral reefs")

print(nemo.name)
print(nemo.species)
print(nemo.length)
print(nemo.color)
print(nemo.habitat)

nemo.swim()
nemo.eat("plankton")
nemo.change_color("Blue")

#세번째 예시
class Virus:
    def __init__(self, name, genome, structure, host):
        self.name = name
        self.genome = genome
        self.structure = structure
        self.host = host

    def infect(self, host):
        print(f"{self.name} is infecting {host}.")

    def replicate(self):
        print(f"{self.name} is replicating.")

    def mutate(self):
        print(f"{self.name} is mutating.")


covid = Virus("COVID-19", "RNA", "Spherical", "Humans")

print(covid.name)
print(covid.genome)
print(covid.structure)
print(covid.host)

covid.infect("humans")
covid.replicate()
covid.mutate()
```
